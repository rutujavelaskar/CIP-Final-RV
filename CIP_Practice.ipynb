```python
# @title Retrieving data
# data from https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.annual_ns_avg.txt
!wget 'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.annual_ns_avg.txt'
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_fwf(
    'HadCRUT.4.6.0.0.annual_ns_avg.txt',
    index_col=0,
    usecols=(0, 1),
    names=['year', 'anomaly'],
    header=None,
)

anomaly = df.loc[1980:2020, 'anomaly'].dropna()

print (anomaly)

plt.plot(anomaly)                # plot the points in the array a
plt.title("Temperature increase 1980-2020")  # Add a title to the figure
plt.xlabel("Years")    # Give a label to the x-axis
plt.ylabel("Deviation from long term average temperature");    # Give a label to the y-axis
```

```python
# @title CIP Final Project:Mammalian Cell Culture Process for Monoclonal Antibody Production
#Data: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4313526/
import matplotlib.pyplot as plt
import pandas as pd
excel_file = '/content/Mammalian Cell Culture Process for Monoclonal Antibody Production Data.xlsx'
df = pd.read_excel(excel_file)
print(df) #prints data file
df = pd.read_excel(excel_file, sheet_name='Sheet2') #functions helps to read the sheet 2 in the excel sheet
print(df)
df = pd.read_excel(excel_file, usecols=['Time','X [106 cells/mL]'])
print(df)

#Processed File

excel_file_processed = '/content/Mammalian Cell Culture Process for Monoclonal Antibody Production Data_processed.xlsx'
df = pd.read_excel(excel_file_processed)
print(df)
df = pd.read_excel(excel_file_processed, sheet_name='Sheet1')
print(df)

df = pd.read_excel(excel_file_processed, skiprows=1) #skips row 1 to as it has non number information
print(df)
# Assuming your Excel file has columns 'X' and 'Y' for plotting
x_values = df[0]
y_values = df[0.09]

# Plotting
plt.figure(figsize=(8, 5)) # Gives the size of plot
plt.plot(x_values, y_values, marker='o', linestyle='-', color='r', label='Data Points')
plt.title('Mammalian Cell Culture')
plt.xlabel('Time')
plt.ylabel('X [106 cells/mL]')
plt.grid(True)
plt.legend()
plt.tight_layout()



# Show the plot
plt.show()

```
